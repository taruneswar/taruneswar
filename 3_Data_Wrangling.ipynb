```python
#Navigating DF

import pandas as pd
url = 'https://raw.githubusercontent.com/chrisalbon/simulated_datasets/master/titanic.csv'
df=pd.read_csv(url)
#df.iloc[0]
df.iloc[1:]
```

```python
#Accessing using the  Index value/name
df=df.set_index(df['Age'])
df.loc[24]
```

```python
#selecting rows based on conditions 
#Note:- the field name is case sensitive

df[ df ['Age'] == 23 ] .head(2)
```

```python
df[ (df ['Age'] == 24)  & (df['Sex'] == 'female')]
```

```python
df['Sex'].replace(['female','women'], ['male','men']).head(5) # we can replace values whatever we want
```

```python
#Renaming cloumns
df.rename(columns={'PClass':'PassengerClass' , 'Sex':'Gender'}).head(2)
```

```python
#unique values 
df['Sex'].value_counts() ## Give number of unique values
```

```python
#Handling missing data

df[df['Age'].isnull()].head(2)
```

```python
#we have to use this to specify the values used to indiacte missing values
import numpy as np
df=pd.read_csv(url,na_values=[np.nan,'NONE',-999])
df
```

```python
#Merging Important

import pandas as pd
e_data = {'id ':['1','2','3','4'],
          'name':['tarun','renu','reddy','eswar']}

data1=pd.DataFrame(e_data,columns = ['id','name'])

e_empoloyees = {'id':['3','4','5','6'],
                        'sales':[12321,12412,12414,4574]}
data2=pd.DataFrame(e_empoloyees,columns = ['id','sales'])
          
pd.merge(data1,data2,on='id',how='outer')
```
